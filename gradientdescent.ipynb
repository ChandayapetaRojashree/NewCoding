```python
def cal_secondDegree(coeffs, x):
  a=coeffs[0]+(x*x)
  b=coeffs[1]*x
  c=coeffs[2]
  y=a+b+c
  return y
hundred_xs=np.random.uniform(-10,10,100)
```

```python
coeffs=[1,0,0]
xs=[]
ys=[]
for x in hundred_xs:
  y=cal_secondDegree(coeffs,x)
  xs.append(x)
  ys.append(y)
```

```python
plt.plot(xs,ys,'g+')
plt.title('orginal data')
plt.show()
```

```python
def eval_secondDegree_jitter(coeffs,x,j):
  a=coeffs[0]*(x*x)
  b=coeffs[1]*x
  c=coeffs[2]
  y=a+b+c
  interval=[y-j,y+j]
  interval_min=interval[0]
  interval_max=interval[1]
  jit_val=random.random()*interval_max
  while jit_val<interval_min:
    jit_val=random.random()*interval_max
  return jit_val
```

```python
xs=[]
ys=[]
for x in hundred_xs:
  y=eval_secondDegree_jitter(coeffs,x,0.1)
  xs.append(x)
  ys.append(y)
```

```python
plt.figure(figsize=(12,6))
plt.plot(xs,ys,'b+')
plt.title('original Data')
plt.show()
```
